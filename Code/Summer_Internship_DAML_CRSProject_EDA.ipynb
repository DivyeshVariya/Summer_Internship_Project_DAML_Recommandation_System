# Exploratory Data Analytics of College Recommendation System

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
new_collegedata=pd.read_csv("new_collegedata.csv")
new_collegedata.head()
```

```python
UG_fee_stream=new_collegedata[['Stream','UG_fee']]
plt.figure(figsize = (16, 8))
plt.bar(UG_fee_stream['Stream'],UG_fee_stream['UG_fee'], color=["green"])
plt.xlabel("Stream")
plt.ylabel("UG_fee")
plt.title("Stream Vs UG_fee")
plt.show()
```

* As , We loop at above bar chart of stream VS UG_fees ,These Chart represent that Under-Graduation fees coresponding Streams. Our Data-set contains various Streams like Engineering , Medical , Commerce etc. with its details.

* Here, We can see that  Medical Stream Education of Under-Graduation too much expenceive compare to other streams and also see that Hotel-Management Stream has minimal fees.

```python
PG_fee_stream=new_collegedata[['Stream','PG_fee']]
plt.figure(figsize = (16, 8))
plt.bar(PG_fee_stream['Stream'],PG_fee_stream['PG_fee'],color='Orange')
plt.xlabel("Stream")
plt.ylabel("PG_fee")
plt.title("Stream Vs PG_fee")
plt.show()
```

* After , Observing above graph of Stream Vs PG_fees , We can see that Post Graduation of Medical Stream fees too much higher as same as Under Graduation fees and Similar , To Streams Vs UG_fees chart that Hotel-Management Stream has minimal fees of Post Graduation. 

```python
import seaborn as sns

stream_UG_PG=new_collegedata[['Stream','PG_fee','UG_fee']]
plt.figure(figsize = (16, 8))
sns.scatterplot(x='UG_fee',y='PG_fee',hue='Stream',data=stream_UG_PG)
plt.title(' Scatter Plot of Stream with UG and PG fees')
# plt.show()
```

* After looking above 2 bar charts of STREAM VS UG_FEE and STREAM VS PG_FEE , We can see that in Scatter plot of Stream with UG_fees and PG_fees , Medical Stream has maximun expences compare to other streams in both Under Graduation and Post Graduation.

* As previously , We was noticed that , also we can saw relationship as same as  here, Hotel-Management Eduaction has minimum expences in both courses UG and PG.

```python
college_count=new_collegedata[['State','College_Name']]
plt.figure(figsize = (16, 8))
sns.countplot(x='State',data=college_count)
plt.xticks(rotation=30,ha='right')
plt.title('State VS College Count')
```

* In these image show the different state wise college count of various streams. In these figure , We can see that Rajasthan State has 175+ colleges of both UG and PG courses in various streams and Andaman State has a few number of colleges. 

* From these observation , We can say that Andaman State's majority of  Students need to go away from his home state to other for study because less number of colleges in state. In case of Rajasthan State , Student has less chances needed for go to other state for study because higher number of colleges with various streams in state.

```python
stream_count_clg=new_collegedata[['Stream','College_Name']]
sns.catplot(x='Stream', kind="count",data=stream_count_clg,height=10)
plt.xticks(rotation=15,ha='center')
plt.title(' College Count VS Stream ')
```

* In College Count VS Stream chart , We can see that Engineering Stream highest number of colleges in India . Similarly , Managemnet , Science , Arts , Pharmacy , commarce and Medical Stream's has good count of colleges in India. Also , We can see that Hotel=Management Stream has less number of colleges in India.

* From Above data , We can say that Engineering , Medical , Science , Arts , Management and Commerce Stream's Student not need worry about colleges because good count of colleges are there , each student can get admission in good college according his/her score.

```python
clg_rating=new_collegedata[['College_Name','Overall_Rating']]
sns.displot(x='Overall_Rating',color=(0,0.75,0.5,1),data=clg_rating,height=10,weights=5)
plt.title(' College Count With Overall Rating ')
```

* Let's , Look at the Overall Rating distribution chart with college count. From these diagram , We can see that majority of colleges has a very nice Rating 7 to 9 . These represent that in India very good colleges are there which best place for higher eduction.

```python
state=new_collegedata['State'].unique()
state.sort()
rating=new_collegedata[['State','College_Name']].groupby('State').count()

explode=[]

for i in range(0,len(state)):
 explode.append(0.2)

plt.figure(figsize = (16, 8))
plt.pie(x=rating.College_Name,labels=state, autopct='%.0f%%',explode=explode,radius=2)
plt.xticks(rotation=15,ha='center')
plt.title(' College Count Per State ')
plt.show()
```

* We look at the above Pie chart of College Count per State , In which we can see that Rajasthan and Haryana State has miximum number of colleges in all over India so that we can say that Rajasthan and haryana State's Students have less chance go out-side from his/her home state for study and also in both state has good enviorment for higher studies.

* Also , We can see the Manipur , Mizorm and Andaman has few colleges in state , It's represent that these are not good place for higher studies.

```python
rating_clg=new_collegedata[['UG_fee','PG_fee','Overall_Rating']]
plt.figure(figsize = (16, 8))
sns.scatterplot(data=rating_clg, x="UG_fee", y="PG_fee", hue='Overall_Rating')
plt.xticks(rotation=15,ha='center')
plt.title(' Course Fees with Overall Rating ')
plt.show()
```

* Above diagram , We can see that as Under Graduation and Post Graduation fees will increases College's Overall Rating also good in number like 9 out of 10 . These relation also say that higher fees of courses college has good enviorment and place for higher eduaction.

```python
rajasthan_clg_stream=new_collegedata['Stream'].unique()
rajasthan_clg_stream.sort()
df=new_collegedata[['State','College_Name','Stream']]
rajasthan_clg=df[df.State=='Rajasthan']
rajasthan_clg_count=rajasthan_clg.groupby('Stream').count()


plt.figure(figsize = (16, 8))
plt.pie(x=rajasthan_clg_count.College_Name,labels=rajasthan_clg_stream, autopct='%.0f%%')
plt.title(' College Count With Stream In Rajasthan State')
plt.show()
```

* As , We observed that Rajasthan State has Maximum number of colleges so we can see in above pie chart represent stream wise college ratio. In Rajasthan State , Maximum colleges of Law and Engineering Stream and also good number of colleges of other streams.
